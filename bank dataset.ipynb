**Analysis of bank dataset by using valuecounts and plot package**

```python
bank_l=[]
for i in bank.select_dtypes(include="object"):
    b=bank[i].value_counts()
    k=b.keys()
    v=b.values
    cols=["labels","count"]
    bank_df=pd.DataFrame(zip(k,v),columns=cols)
    bank_l.append(bank_df)
    path=r"C:\Users\HP\valuecount"
    bank_df.to_csv(f"{path}\\{i}.csv",index=False)
```

```python
l1=list(bank.select_dtypes(include="object").columns)
```

```python
import matplotlib.pyplot as plt
for i,j in zip(bank_l,l1):
    plt.figure(figsize=(19,8))
    plt.bar("labels","count",data=i)
    plt.title(f"bar graph--- {j}")
    plt.xlabel("class")
    plt.ylabel("frequency")
    path=r"C:\Users\HP\valuecount"
    plt.savefig(f"{path}\\{j}.jpg")
    
    plt.show()

```
