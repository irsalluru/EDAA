```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
bank=pd.read_csv(r"C:\Users\HP\EDA\dataset\bank.csv",sep=";")
bank
```

```python
num_cols=bank.select_dtypes(exclude="object").columns
num_cols
```

```python
new_dir=os.path.join(os.getcwd(),"bank_histogram")
os.makedirs(new_dir)
```

```python

for i in num_cols:
    plt.hist(bank[i])
    plt.title(i)
    plt.xlabel("class interval")
    plt.ylabel("frequency")
    plt.savefig(f"{new_dir}\\ {i}.jpg")
    
    plt.show()
```
