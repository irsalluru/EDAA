```python
visa_df=pd.read_csv(r"C:\Users\HP\EDA\dataset\Visadataset.csv")
```

```python
cat_cols=visa_df.select_dtypes(include="object").columns
num_cols=visa_df.select_dtypes(exclude="object").columns
```

```python
#Step1: read the package
from sklearn.preprocessing import LabelEncoder
#step2:save the package/ call the method
le=LabelEncoder()
#step3: Appply fit ransform on data
for i in cat_cols:
    visa_df[i]=le.fit_transform(visa_df[i])
visa_df
```

**fit and transform**
- fit means develop the logic
- transform means process the logic
- if we observe in map method first we developed dictionary lo;gic

- developing the dictionary is akind of fit
- apply the dictionary to column means we are transforming the data cat to num

**np.where**
- using np.where also we can chage the Categorical to numerical data
- but np.where is only for binary condition
- np.where is same as if and else
- if condition is true : will aply True value
- if condition is false: will apply False value
- A column should have only 2 values


```python
visa_df=pd.read_csv(r"C:\Users\HP\EDA\dataset\Visadataset.csv")
visa_df["case_status"]=np.where(visa_df["case_status"]=="Certified",0,1)
visa_df
```
